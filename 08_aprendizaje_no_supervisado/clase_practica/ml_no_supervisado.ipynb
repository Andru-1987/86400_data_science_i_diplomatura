### ¿Qué es el Aprendizaje No Supervisado?

El objetivo principal es **encontrar patrones, estructuras o relaciones ocultas en datos que no tienen etiquetas**. A diferencia del aprendizaje supervisado, no hay una "respuesta correcta" predefinida. El modelo debe explorar los datos por sí mismo para descubrir información valiosa.


### 1. Clustering (Agrupamiento)

Es la técnica que mencionas. Su objetivo es agrupar datos en "clusters" o conjuntos de tal manera que:
*   Los puntos dentro de un mismo grupo sean muy similares entre sí.
*   Los puntos de grupos diferentes sean muy distintos.

**Aplicaciones del Clustering:**
*   **Segmentación de clientes:** Agrupar clientes por comportamiento de compra, demografía, etc., para marketing dirigido.
*   **Agrupación de documentos:** Organizar artículos de noticias o documentos de investigación por temas.
*   **Detección de anomalías:** Identificar transacciones bancarias fraudulentas (son "clusters" muy pequeños y atípicos).
*   **Bioinformática:** Clasificar diferentes tipos de células o genes con patrones de expresión similares.

---

### 2. Reducción de Dimensionalidad

Esta técnica busca **simplificar los datos** sin perder su información esencial. Se usa cuando los datos tienen un número muy alto de características (dimensiones), lo que puede ser computacionalmente costoso y difícil de visualizar.

**Aplicaciones de la Reducción de Dimensionalidad:**
*   **Compresión de imágenes y datos:** Técnicas como PCA se usan para reducir el tamaño de archivos de imagen manteniendo la mayor parte de la calidad.
*   **Visualización de datos:** Proyectar datos de alta dimensión (como datos genómicos) en 2D o 3D para poder graficarlos y entenderlos.
*   **Preprocesamiento de datos:** Mejorar el rendimiento de otros modelos de Machine Learning (como los supervisados) eliminando el "ruido" y las características redundantes.

---

### 3. Reglas de Asociación

Esta técnica descubre **relaciones interesantes entre variables** en bases de datos grandes. La pregunta que responde es: "Si un evento ocurre, ¿qué otro evento probablemente ocurrirá con él?".

**Aplicaciones de las Reglas de Asociación:**
*   **Análisis de mercado (Market Basket Analysis):** El ejemplo clásico: "Los clientes que compran pañales, a menudo también compran cerveza". Se usa para optimizar la disposición de productos en un supermercado o para hacer recomendaciones de productos complementarios.
*   **Sistemas de recomendación:** Sugerir música, películas o libros basándose en lo que otros usuarios con gustos similares han consumido.
*   **Análisis de síntomas médicos:** Encontrar combinaciones de síntomas que suelen aparecer juntos en ciertas enfermedades.

---

### 4. Modelado de Temas (Topic Modeling)

Es una técnica específica para texto que **descubre "temas" abstractos** que ocurren en una colección de documentos.

**Aplicaciones del Modelado de Temas:**
*   **Análisis de opiniones (Sentiment Analysis):** Descubrir los temas principales sobre los que hablan los clientes en sus reseñas.
*   **Organización de grandes colecciones de documentos:** Automáticamente etiquetar y categorizar artículos de noticias, patentes o correos electrónicos por su contenido temático.
*   **Investigación de mercado:** Analizar grandes volúmenes de comentarios en redes sociales para entender qué temas son tendencia respecto a una marca.

### **1. Clustering (Agrupamiento)**

**Algoritmos más comunes:**

*   **K-Means (K-Medias):**
    *   **Cómo funciona:** Divide los datos en `k` clusters. Busca posicionar `k` centroides (el punto central de cada cluster) de manera que la distancia de cada punto a su centroide más cercano sea mínima.
    *   **Pros:** Muy rápido y eficiente para grandes volúmenes de datos.
    *   **Contras:** Hay que especificar `k` de antemano, sensible a valores atípicos y asume clusters de forma esférica.

*   **DBSCAN (Density-Based Spatial Clustering of Applications with Noise):**
    *   **Cómo funciona:** Agrupa puntos que están muy juntos (alta densidad) y marca como "ruido" los puntos que quedan aislados en regiones de baja densidad.
    *   **Pros:** No necesita que especifiques el número de clusters. Puede encontrar clusters de cualquier forma y detecta valores atípicos.
    *   **Contras:** Menos efectivo cuando la densidad de los clusters es muy variable.

*   **Hierarchical Clustering (Clustering Jerárquico):**
    *   **Cómo funciona:** Crea un árbol de clusters (dendrograma). Puede ser "agglomerative" (cada punto empieza como un cluster y se van uniendo) o "divisive" (todos los puntos empiezan en un cluster y se van dividiendo).
    *   **Pros:** No necesita un `k` inicial y el dendrograma es muy interpretable.
    *   **Contras:** Computacionalmente costoso para grandes conjuntos de datos.

---

### **2. Reducción de Dimensionalidad**

**Algoritmos más comunes:**

*   **PCA (Análisis de Componentes Principales):**
    *   **Cómo funciona:** Encuentra las direcciones (componentes principales) en los datos que tienen la mayor varianza y proyecta los datos en estas nuevas direcciones. Son combinaciones lineales de las características originales.
    *   **Pros:** Muy eficiente y bien entendido. Excelente para eliminar correlaciones entre características.
    *   **Contras:** Es un método lineal y puede perder estructuras no lineales complejas.

*   **t-SNE (t-Distributed Stochastic Neighbor Embedding):**
    *   **Cómo funciona:** Se centra en preservar la estructura local de los datos. Intenta que puntos que son vecinos en el espacio original sigan siendo vecinos en el espacio de baja dimensión.
    *   **Pros:** Excelente para visualización, ya que revela clusters muy claros.
    *   **Contras:** Muy lento. Los resultados pueden variar entre ejecuciones y la escala de los ejes en la visualización no tiene significado.

*   **UMAP (Uniform Manifold Approximation and Projection):**
    *   **Cómo funciona:** Similar a t-SNE, pero con un enfoque teórico diferente. También preserva la estructura local pero es mucho más rápido y tiende a preservar mejor la estructura global de los datos.
    *   **Pros:** Muy rápido y potente. Se ha vuelto extremadamente popular para la visualización de datos de alta dimensión (como en genómica).
    *   **Contras:** Al igual que t-SNE, tiene hiperparámetros que pueden afectar el resultado.

---

### **3. Reglas de Asociación**

**Algoritmo más común:**

*   **Apriori:**
    *   **Cómo funciona:** Encuentra conjuntos de elementos frecuentes (itemsets) en los datos y luego genera reglas a partir de ellos (ej: "Si {pañales, leche} entonces {cerveza}"). Utiliza una propiedad de "anti-monotonía": si un conjunto es frecuente, todos sus subconjuntos también lo son.
    *   **Pros:** Fácil de entender e implementar.
    *   **Contras:** Puede ser muy lento para un número grande de elementos, ya que requiere escanear la base de datos muchas veces.

*   **FP-Growth (Frequent Pattern Growth):**
    *   **Cómo funciona:** Es una mejora sobre Apriori. Construye una estructura de árbol (FP-tree) para comprimir la base de datos, lo que le permite encontrar los conjuntos frecuentes de manera más eficiente.
    *   **Pros:** Mucho más rápido que Apriori.
    *   **Contras:** La implementación es más compleja.

---

### **4. Modelado de Temas (Topic Modeling)**

**Algoritmo más común:**

*   **LDA (Latent Dirichlet Allocation):**
    *   **Cómo funciona:** Es un modelo generativo probabilístico. Asume que cada documento es una mezcla de varios temas y que cada tema es una distribución de probabilidad sobre un conjunto de palabras.
    *   **Pros:** Es el algoritmo más popular y establecido para esta tarea. Los resultados son bastante interpretables.
    *   **Contras:** Requiere que especifiques el número de temas de antemano. Puede ser sensible a la preparación del texto (stopwords, stemming, etc.).

### Clusterizacion

```python
def calcular_estadisticas_equipos(df):
    """
    Calcula STATS completas para cada equipo
    separando rendimiento como local y visitante
    """
    equipos_unicos = pd.concat([df['equipo_local'], df['equipo_visitante']]).unique()
    estadisticas = []
    
    for equipo in equipos_unicos:
        # Filtrar partidos donde el equipo participa -> Separacion de visitantes vs locales
        partidos_local = df[df['equipo_local'] == equipo]
        partidos_visitante = df[df['equipo_visitante'] == equipo]
        total_partidos = len(partidos_local) + len(partidos_visitante)
        
        #  RENDIMIENTO OFENSIVO 
        goles_local = partidos_local['goles_local'].sum()
        goles_visitante = partidos_visitante['goles_visitante'].sum()
        goles_totales = goles_local + goles_visitante
        
        #  RENDIMIENTO DEFENSIVO 
        goles_recibidos_local = partidos_local['goles_visitante'].sum()
        goles_recibidos_visitante = partidos_visitante['goles_local'].sum()
        goles_recibidos_totales = goles_recibidos_local + goles_recibidos_visitante
        
        #  RESULTADOS 
        victorias_local = len(partidos_local[partidos_local['resultado'] == 'L'])
        victorias_visitante = len(partidos_visitante[partidos_visitante['resultado'] == 'V'])
        empates_local = len(partidos_local[partidos_local['resultado'] == 'E'])
        empates_visitante = len(partidos_visitante[partidos_visitante['resultado'] == 'E'])
        
        victorias_totales = victorias_local + victorias_visitante
        empates_totales = empates_local + empates_visitante
        derrotas_totales = total_partidos - victorias_totales - empates_totales
        
        stats = {
            'equipo': equipo,
            
            #  STATS GENERALES 
            'partidos_totales': total_partidos,
            'partidos_local': len(partidos_local),
            'partidos_visitante': len(partidos_visitante),
            
            #  RENDIMIENTO OFENSIVO 
            'goles_totales': goles_totales,
            'goles_promedio': goles_totales / total_partidos if total_partidos > 0 else 0,
            'goles_local_total': goles_local,
            'goles_local_promedio': goles_local / len(partidos_local) if len(partidos_local) > 0 else 0,
            'goles_visitante_total': goles_visitante,
            'goles_visitante_promedio': goles_visitante / len(partidos_visitante) if len(partidos_visitante) > 0 else 0,
            
            #  RENDIMIENTO DEFENSIVO 
            'goles_recibidos_totales': goles_recibidos_totales,
            'goles_recibidos_promedio': goles_recibidos_totales / total_partidos if total_partidos > 0 else 0,
            'diferencia_goles': goles_totales - goles_recibidos_totales,
            
            #  RESULTADOS 
            'victorias_totales': victorias_totales,
            'empates_totales': empates_totales,
            'derrotas_totales': derrotas_totales,
            'porcentaje_victorias': (victorias_totales / total_partidos) * 100 if total_partidos > 0 else 0,
            'porcentaje_empates': (empates_totales / total_partidos) * 100 if total_partidos > 0 else 0,
            'porcentaje_derrotas': (derrotas_totales / total_partidos) * 100 if total_partidos > 0 else 0,
            'puntos_totales': (victorias_totales * 3) + empates_totales,  # Asumiendo 3 puntos por victoria
            
            #  STATS AVANZADAS (promedios) 
            'posesion_promedio': (
                partidos_local['posesion_local'].mean() + 
                partidos_visitante['posesion_visitante'].mean()
            ) / 2 if len(partidos_local) > 0 and len(partidos_visitante) > 0 else 0,
            
            'tiros_arco_promedio': (
                partidos_local['tiros_arco_local'].mean() + 
                partidos_visitante['tiros_arco_visitante'].mean()
            ) / 2 if len(partidos_local) > 0 and len(partidos_visitante) > 0 else 0,
            
            'faltas_promedio': (
                partidos_local['faltas_local'].mean() + 
                partidos_visitante['faltas_visitante'].mean()
            ) / 2 if len(partidos_local) > 0 and len(partidos_visitante) > 0 else 0,
            
            'amarillas_promedio': (
                partidos_local['amarillas_local'].mean() + 
                partidos_visitante['amarillas_visitante'].mean()
            ) / 2 if len(partidos_local) > 0 and len(partidos_visitante) > 0 else 0,
            
            #  CARACTERÍSTICAS DEL PLANTEL 
            'valor_mercado': partidos_local['valor_mercado_local'].iloc[0] if len(partidos_local) > 0 else partidos_visitante['valor_mercado_visitante'].iloc[0],
            'altura_media': partidos_local['altura_media_local'].iloc[0] if len(partidos_local) > 0 else partidos_visitante['altura_media_visitante'].iloc[0],
            'edad_media': partidos_local['edad_media_local'].iloc[0] if len(partidos_local) > 0 else partidos_visitante['edad_media_visitante'].iloc[0],
            'proporcion_zurdos': partidos_local['proporcion_zurdos_local'].iloc[0] if len(partidos_local) > 0 else partidos_visitante['proporcion_zurdos_visitante'].iloc[0],
            
            #  EFECTIVIDAD 
            'efectividad_ataque': goles_totales / (
                partidos_local['tiros_arco_local'].sum() + 
                partidos_visitante['tiros_arco_visitante'].sum()
            ) if (partidos_local['tiros_arco_local'].sum() + partidos_visitante['tiros_arco_visitante'].sum()) > 0 else 0,
            
            'efectividad_defensa': 1 - (goles_recibidos_totales / (
                partidos_local['tiros_arco_visitante'].sum() + 
                partidos_visitante['tiros_arco_local'].sum()
            )) if (partidos_local['tiros_arco_visitante'].sum() + partidos_visitante['tiros_arco_local'].sum()) > 0 else 0
        }
        
        estadisticas.append(stats)
    
    return pd.DataFrame(estadisticas)
```

```python
# Calcular estadísticas
df_equipos = calcular_estadisticas_equipos(df)
df_equipos.head()
```

```python
# Columnas recomendadas para clustering

columnas_clustering = [
    # Rendimiento
    'goles_promedio',
    'goles_recibidos_promedio', 
    'diferencia_goles',
    'porcentaje_victorias',
    'puntos_totales',
    
    # Estilo de juego
    'posesion_promedio',
    'tiros_arco_promedio',
    'faltas_promedio',
    'amarillas_promedio',
    
    # Efectividad
    'efectividad_ataque',
    'efectividad_defensa',
    
    # Características del plantel
    'valor_mercado',
    'altura_media',
    'edad_media'
]

# Preparar datos para clustering
X_equipos = df_equipos[columnas_clustering].fillna(0)

print(f"Dataset para clustering: {X_equipos.shape}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.decomposition import PCA
```

```python
# Preparar datos
X = df_equipos[columnas_clustering].fillna(0)

# Escalar los datos (MUY IMPORTANTE para clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Agregar nombres de equipos para referencia
equipos = df_equipos['equipo'].values
```

# K-Means

```python
# Determinar número óptimo de clusters con método del codo
# Inercia = Suma de las distancias al cuadrado de cada punto a su centroide más cercano

inertia = []
k_range = range(2, 20)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Graficar método del codo
plt.figure(figsize=(12, 7))
plt.plot(k_range, inertia, 'bo-', linewidth=2, markersize=8)

# Agregar etiquetas con los valores de inercia
for k, inert in zip(k_range, inertia):
    plt.annotate(f'{inert:.1f}', 
                (k, inert),
                textcoords="offset points",
                xytext=(0,10),
                ha='center',
                fontsize=9,
                alpha=0.7)

plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.title('Método del Codo para K-Means')
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Calcular la diferencia porcentual de inercia
diferencias = []

for i in range(1, len(inertia)):
    reduccion_porcentual = ((inertia[i-1] - inertia[i]) / inertia[i-1]) * 100
    diferencias.append(reduccion_porcentual)
print(f"Diferencias de {len(diferencias)}: ", diferencias)

# Encontrar donde la reducción se vuelve menos significativa
umbral_reduccion = 10  # 10% de reducción como umbral
for i, reduccion in enumerate(diferencias):
    if reduccion < umbral_reduccion:
        k_optimo = i + 2  # +2 porque empezamos desde k=2
        break

print(f"K óptimo: {k_optimo}")
```

```python
# Aplicar K-Means con número óptimo (ejemplo: 4 clusters)
kmeans = KMeans(n_clusters=k_optimo, random_state=42, n_init=10)
clusters_kmeans = kmeans.fit_predict(X_scaled)

# Agregar clusters al DataFrame
df_equipos['cluster_kmeans'] = clusters_kmeans
```

##  **Qué son los parámetros de DBSCAN:**

### **1. `eps` (épsilon) - Radio de búsqueda**
- **Qué es**: La distancia máxima entre dos puntos para ser considerados "vecinos"
- **Analogía**: El radio de tu "círculo de amistad"
- **Efecto**:
  - `eps` muy pequeño → Muchos clusters pequeños, muchos puntos como ruido (-1)
  - `eps` muy grande → Pocos clusters grandes, todo se une

### **2. `min_samples` - Mínimo de vecinos**
- **Qué es**: Número mínimo de puntos requeridos para formar un cluster denso
- **Analogía**: Mínimo de amigos necesarios para formar un grupo
- **Efecto**:
  - `min_samples` pequeño → Clusters más pequeños, más sensibilidad a ruido
  - `min_samples` grande → Clusters más grandes, más robusto a outliers

eps muy pequeño → Muchos clusters pequeños, muchos puntos como ruido (-1)
eps muy grande → Pocos clusters grandes, todo se une

```python
# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=3)
clusters_dbscan = dbscan.fit_predict(X_scaled)

# Agregar clusters al DataFrame
df_equipos['cluster_dbscan'] = clusters_dbscan

print("Distribución de clusters DBSCAN:")
print(pd.Series(clusters_dbscan).value_counts().sort_index())
```

```python
# Parámetros conservadores para empezar
parametros_recomendados = [

    (0.001, 3),   # Ultra sensible - clusters muy pequeños
    (0.5, 3),   # Más sensible - clusters pequeños
    (0.7, 4),   # Balanceado - buen punto de partida
    (0.9, 5),   # Más conservador - clusters más grandes
    (1.1, 6)    # Muy conservador - pocos clusters grandes
]

for eps, min_samples in parametros_recomendados:
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    clusters = dbscan.fit_predict(X_scaled)
    
    n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
    n_ruido = list(clusters).count(-1)
    
    print(f"eps={eps}, min_samples={min_samples}:")
    print(f"  - Clusters encontrados: {n_clusters}")
    print(f"  - Puntos como ruido: {n_ruido} ({n_ruido/len(X_scaled)*100:.1f}%)")
    print(f"  - Distribución: {pd.Series(clusters).value_counts().sort_index()}")
    print()
```

## Jerarquico

```python
# Aplicar clustering jerárquico
Z = linkage(X_scaled, method='ward')

# Graficar dendrograma
plt.figure(figsize=(15, 8))
dendrogram(Z, labels=equipos, leaf_rotation=90, leaf_font_size=10)
plt.title('Dendrograma - Clustering Jerárquico')
plt.xlabel('Equipos')
plt.ylabel('Distancia')
plt.tight_layout()
plt.show()

# Cortar el dendrograma para obtener clusters
clusters_hierarchical = fcluster(Z, t=4, criterion='maxclust')
df_equipos['cluster_hierarchical'] = clusters_hierarchical - 1  # Ajustar a base 0
```

```python
# Reducción de dimensionalidad para visualización 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Crear DataFrame para visualización
df_viz = pd.DataFrame({
    'equipo': equipos,
    'PCA1': X_pca[:, 0],
    'PCA2': X_pca[:, 1],
    'kmeans': clusters_kmeans,
    'dbscan': clusters_dbscan,
    'hierarchical': clusters_hierarchical - 1
})

# Función para graficar con nombres
def plot_clusters(df, cluster_col, title, figsize=(15, 10)):
    plt.figure(figsize=figsize)
    
    # Colores para clusters
    unique_clusters = df[cluster_col].unique()
    colors = plt.cm.Set3(np.linspace(0, 1, len(unique_clusters)))
    
    for i, cluster in enumerate(unique_clusters):
        cluster_data = df[df[cluster_col] == cluster]
        plt.scatter(cluster_data['PCA1'], cluster_data['PCA2'], 
                   c=[colors[i]], label=f'Cluster {cluster}', s=100, alpha=0.7)
        
        # Añadir nombres de equipos
        for _, row in cluster_data.iterrows():
            plt.annotate(row['equipo'], 
                        (row['PCA1'], row['PCA2']),
                        xytext=(5, 5), 
                        textcoords='offset points',
                        fontsize=9,
                        alpha=0.8)
    
    plt.xlabel('Componente Principal 1')
    plt.ylabel('Componente Principal 2')
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Graficar los tres métodos
plot_clusters(df_viz, 'kmeans', 'K-Means Clustering de Equipos')
plot_clusters(df_viz, 'dbscan', 'DBSCAN Clustering de Equipos')
plot_clusters(df_viz, 'hierarchical', 'Clustering Jerárquico de Equipos')
```

```python
# Función para analizar características de cada cluster
def analizar_clusters(df_equipos, cluster_col):
    analysis = df_equipos.groupby(cluster_col).agg({
        'goles_promedio': 'mean',
        'goles_recibidos_promedio': 'mean',
        'porcentaje_victorias': 'mean',
        'posesion_promedio': 'mean',
        'valor_mercado': 'mean',
        'equipo': 'count'
    }).round(3).sort_values(by="porcentaje_victorias")
    
    analysis = analysis.rename(columns={'equipo': 'num_equipos'})
    return analysis.to_string()

print(" K-MEANS CLUSTERS ")
print(analizar_clusters(df_equipos, 'cluster_kmeans'))
print("\n DBSCAN CLUSTERS ")
print(analizar_clusters(df_equipos, 'cluster_dbscan'))
print("\n HIERARCHICAL CLUSTERS ")
print(analizar_clusters(df_equipos, 'cluster_hierarchical'))

# Mostrar equipos por cluster para K-Means (ejemplo)
print("\n EQUIPOS POR CLUSTER (K-MEANS) ")
for cluster in sorted(df_equipos['cluster_kmeans'].unique()):
    equipos_cluster = df_equipos[df_equipos['cluster_kmeans'] == cluster]['equipo'].tolist()
    print(f"Cluster {cluster} ({len(equipos_cluster)} equipos): {', '.join(equipos_cluster)}")
```
