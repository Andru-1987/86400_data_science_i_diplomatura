```python
escuelita_recuperatorios = {
    "edades" : [4,12,10,9],
    "calificaciones" : [None, None, 9,8]
}
```

```python
recuperatorios = pd.DataFrame(escuelita_recuperatorios)
```

```python
recuperatorios.info()
```

```python
calificaciones_mean = recuperatorios.calificaciones.mean()
print("promedio de calificacion\t", calificaciones_mean)
```

```python
recuperatorios.calificaciones.fillna(calificaciones_mean, inplace=True)
```

```python
recuperatorios
```

```python
df_escuela_nancy.info()
```

```python
recuperatorios.info()
```

```python
# UNION ALL | UNION

escuelita = pd.concat([ df_escuela_nancy, recuperatorios]).reset_index(drop=True)
escuelita.info()
```

```python
escuelita.groupby(["edades"])["calificaciones"].mean()
```

```python
escuelita.groupby("edades").agg(
    {
        "calificaciones": ["mean","min","max","count"],
        "aprobado": "mean"
    }
)
```
